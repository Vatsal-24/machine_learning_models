#**Linear Regression on diabetes dataset**


```python
# Importing dataset
from sklearn import datasets

#importing other libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = datasets.load_diabetes()
```

```python
# Arranging data in tabular format from the dictionary
dataset=pd.DataFrame(data.data)
#Adding column names
dataset.columns=data.feature_names
dataset.head()
```

```python
## Independent features and dependent features
X=dataset
Y=data.target
```

```python
## train test split 
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, train_size=0.9, random_state=99, shuffle=False)
```

```python
# ## standardizing the dataset
# from sklearn.preprocessing import StandardScaler
# scaler = StandardScaler()
# X_train=scaler.fit_transform(X_train)
# X_test=scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
#cross validation
regression=LinearRegression()
regression.fit(X_train,Y_train)
```

```python
#prediction
prediction=regression.predict(X_test)
prediction
```

```python
result = pd.DataFrame({'Actual': Y_test, 'Predict' : prediction})
result
```

```python
# we will check the accuracy

print('coefficient', regression.coef_)
print('intercept', regression.intercept_)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
# mean_squared_error
mean_squared_error(Y_test,prediction)
```

```python
# r2 score
r2_score(Y_test,prediction)
```
